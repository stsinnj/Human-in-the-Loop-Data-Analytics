# Change date format  
Because in raw data, the date format of some datasets will be different, so I process the time data here, so that the format is yyyy-mm-dd.  
(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)

```python
"(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)"
#change foramt to datetime
data1['Start'] = pd.to_datetime(data1['Start'])
#change the datetime format as yyyy-mm-dd
data1['Start'] = data1['Start'].dt.strftime('%Y/%m/%d')
#group by start time and count the steps
day_data1 = data1.groupby(['Start']).sum()
#show the data
day_data1
```

```python
"(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)"
#change foramt to datetime
data2['Start'] = pd.to_datetime(data2['Start'])
#change the datetime format as yyyy-mm-dd
data2['Start'] = data2['Start'].dt.strftime('%Y/%m/%d')
#group by start time and count the steps
day_data2 = data2.groupby(['Start']).sum()
#show the data
day_data2
```

Because after the above data processing, the data set is missing index, so we need to add index.  
(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)

```python
"(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)"
# add index
day_data1 = day_data1.reset_index(drop=False)
day_data1
```

```python
"(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)"
# add index
day_data2 = day_data2.reset_index(drop=False)
day_data2
```

Because the weekly data needs to be obtained, the resample function can be used to easily integrate the weekly data, and then use the sum to find the number of steps taken in a week.  
(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)

```python
"(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)"
#first change the data type to datetime
day_data1['Start'] = pd.to_datetime(day_data1['Start'])
#use resample to integrate the data into weekly
week_data1 = day_data1.resample('W', on="Start").sum()
#show the data
week_data1
```

```python
"(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)"
#first change the data type to datetime
day_data2['Start'] = pd.to_datetime(day_data2['Start'])
#use resample to integrate the data into weekly
week_data2 = day_data2.resample('W', on="Start").sum()
#show the data
week_data2
```

After getting the data for each week, we still don't know which week the data is, so I added numbers to record each week.  
(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)

```python
"(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)"
#added numbers to record each week
week_data1['weeks'] = range(len(week_data1))
#show the data
week_data1
```

```python
"(beginning date：16/10/2022, end date： 16/10/2022, author: Tiannan Chen)"
#added numbers to record each week
week_data2['weeks'] = range(len(week_data2))
#show the data
week_data2
```
