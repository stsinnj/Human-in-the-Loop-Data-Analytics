### Barplot
This is a barplot of two weeks step data with the red line as the strat of second week.

```python
# #started 2022/9/20 finished 2022/9/20 by Sheng
#barplot
ax = m.loc[0:335, ['Steps (count)']].plot(kind='bar', figsize=(100, 4), title='Barplot between two weeks', rot=0)
_ = ax.vlines(x=[167], ymin=0, ymax=8000, colors='r')
print(ax)
```

The daily pattern could be clearly observed. With higher step count at day time. For the week's comparison, the steps variation seems similar, with two days of high step counts each week, and similar on the other days.
