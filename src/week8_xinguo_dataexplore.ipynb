Below is process of exploring first day of data

```python
frist_day = raw_data[:24]
frist_day
```

Sort the first day' data by number of steps. 
We could find:
- the participant had more than six thousand steps from 18:00 to 19:00
- The participant's activity level has a large difference among several hours.
- Three hours' steps are more than a thousand.

```python
frist_day.sort_values(by=['Steps (count)'], ascending=False)
```

Below is the visualization of the participant's activity level. Barplot could represent the amount of steps in different time. Line plot is not suitable here because it mainly shows the trend of steps.  

```python
x = range(1, 25)

plt.bar(
    x,
    frist_day['Steps (count)']
)
plt.xlabel("Hours")
plt.ylabel("Steps")
plt.title("Steps track of participant-8 on 01-Jan-2022")
plt.show()
```
